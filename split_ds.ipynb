```python
data_list = open("data/corpus_paramopama+second_harem.txt", encoding='utf-8').readlines()
result = []
```

```python
data_list
```

```python
def split_sentences(data):
    result = []
    
    sentence = []
    for text in data:
        sentence.append(text)
        if text == '\n':
            result.append(sentence)
            sentence = []
    
    return result
```

```python
result = split_sentences(data_list)
```

```python
for r in result:
    print(r)
    break
```

```python
from numpy.random import shuffle
```

```python
shuffle(result)
sep = int(len(result)*0.8)

train_data = result[:sep]
eval_data = result[sep:]
```

```python
def data_to_txt(data, file_name):
    result = []
    for sent in data:
        result += sent
        
    fp = open(file_name, 'w',encoding="utf-8")
    fp.writelines(result)
    return result

data_to_txt(train_data, "data/custom_train.txt")
data_to_txt(eval_data, "data/custom_eval.txt")
```
